```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#bringing in and cleaning up data
df1 = pd.read_csv('interest survey responses.csv')
df = df1.iloc[:, [2, 3, 4, 5, 8, 9, 13, 14, 15, 16]]
df = df.rename(columns={'What is your class standing for the 2020-2021 school year?': 'standing', 'I am an...': 'location', 'Which statements below best describe you? (Select all that apply)': 'engagement', 'In what ways do you wish you had more support during your first year?': 'support', 'Rate your level of agreement with the following statements: [I had a hard time finding a group that I feel I belong to.]': 'belong', 'Rate your level of agreement with the following statements: [I find it hard to make friends at UW.]': 'friends', 'Rate your level of agreement with the following statements: [I initially felt lost coming when I came to UW.]': 'feltlost', 'Rate your level of agreement with the following statements: [I find it easy to form study groups.]': 'formstudygroup', 'Rate your level of agreement with the following statements: [I have a lot of friends from class or study groups.]': 'friendsfrom', 'Unnamed: 4': 'residence'})
```

```python
#people who responded 'none of the above' to the question 'which statements below best describe you?'
noneng = df[df.engagement == 'None of the above']
noneng = noneng[['standing', 'location']]

grouped = noneng.groupby(['standing', 'location']).size()
grouped.unstack().plot(kind='bar', stacked=True)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
#people that responded 'slightly agree' to the question 'i had a hard time finding a group that i feel i belong to'
belong = df[df.belong == 'Slightly Agree']
belong = belong[['standing', 'location']]

grouped = belong.groupby(['standing', 'location']).size()
grouped.unstack().plot(kind='bar', stacked=True)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```
